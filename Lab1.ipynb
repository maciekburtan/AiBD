Wskazówki: Deklaracja funkcji w Python.

W Pythonie bloki kodu (w tym również funkcje) są wyróżniane za pomocą wcięć. Do wizualizacji wyników w Pythonie używa się pakietu Matplotlib.

W pliku mają być zawarte: imię, nazwisko, opis wykonywanego ćwiczenia - formatowanie markdown w poszczególnych komórkach.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**2+5
x1 = np.linspace(-1,1)
x2 = np.linspace(-6,6)
x3 = np.linspace(0,5)
y1 = f(x1)
y2 = f(x2)
y3 = f(x3)
```

$x>-1 oraz x<1$

```python
plt.plot(x1,y1)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title("$x>-1 & x<1$")
plt.legend(['f(x)'])
plt.show
```

$x>-6 oraz x<6$

```python
plt.plot(x2,y2)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title("$x>-6 & x<6$")
plt.legend(['f(x)'])
plt.show
```

$x>0 oraz x<5$

```python
plt.plot(x3,y3)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title("$x>0 & x<5$")
plt.legend(['f(x)'])
plt.show
```
