Ćwiczenie 2 - TIER protocol i tidy data
Celem ćwiczenia jest pobranie i uporządkowanie przydzielonego zbioru danych zgodnie z protokołem TIER i zasadami "tidy data".

Dla przypomnienia:

W uporządkowanych danych:

-Każda zmienna tworzy kolumnę.

-Każda obserwacja tworzy rząd.

-Każdy typ jednostki obserwacyjnej tworzy tabelę.


tb.csv - dane o gruźlicy w różnych grupach pacjentów (tabela koduje jednocześnie informacje o wieku i o płci w kolumnach i zawiera dużo pustych miejsc)

```python
import pandas as pd
```

```python
tb_original = pd.read_csv('../Original Data/tb.csv')
```

```python
tb_original
```

```python
tb_copy = tb_original.copy()
```

```python
columns_names = tb_copy.filter(regex='new_sp').columns
```

```python
columns_names
```

```python
df = pd.melt(tb_original, id_vars=["iso2","year"], value_name="cases", var_name="sex_and_age")

tmp_df = df["sex_and_age"].str.extract("(\D)(\d+)(\d{2})")    

tmp_df.columns = ["sex", "age_lower", "age_upper"]

tmp_df["age"] = tmp_df["age_lower"] + "-" + tmp_df["age_upper"]

df = pd.concat([df, tmp_df], axis=1)

df = df.drop(['sex_and_age',"age_lower","age_upper"], axis=1)
df = df.dropna()
df = df.sort_values(by=["iso2", "year", "sex", "age"],ascending=True)
df.head(10)
```

```python
df.to_csv('../Analysis Data/transformed_data.csv', encoding='utf-8', index=False)
```
